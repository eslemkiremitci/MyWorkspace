# **Görüntü Manipülasyonu: scipy.ndimage**

scipy.ndimage, n boyutlu dizileri (örneğin, görüntüler) işlemek için araçlar sağlar. Görüntüler üzerinde çeşitli geometrik dönüşümleri kolayca uygulayabilirsiniz.

```python
from scipy.datasets import face  # Yeni modülden örnek görüntü yüklemek için
from scipy.ndimage import shift, rotate, zoom  # Dönüşüm fonksiyonları
import matplotlib.pyplot as plt  # Görselleştirme için

# Örnek görüntüyü yükle
face_image = face(gray=True)  # Gri tonlamalı görüntü

# Görüntüyü görselleştir
plt.imshow(face_image, cmap='gray')
plt.axis('off')
plt.show()

```

# Görüntü Üzerinde Geometrik Dönüşümler

```python
shifted_face = shift(face_image, (50, 50))  # Görüntüyü (50, 50) piksel kaydır
shifted_face2 = shift(face_image, (50, 50), mode="nearest")  # Kenarları "nearest" yöntemiyle doldur
rotated_face = rotate(face_image, 30)  # Görüntüyü 30 derece döndür
cropped_face = face_image[50:-50, 50:-50]  # Görüntüyü kırp
zoomed_face = zoom(face_image, 2)  # Görüntüyü 2 kat büyüt
```

# Görüntüleri Görselleştirme

```python
plt.figure(figsize=(15, 3))  # Tüm grafikleri içerecek bir figür oluştur

# 1. Kaydırılmış Görüntü
plt.subplot(151)  # 1. pozisyondaki subplot
plt.imshow(shifted_face, cmap=plt.cm.gray)  # Kaydırılmış görüntüyü çiz
plt.axis('off')  # Eksenleri gizle

# 2. "Nearest" ile Kaydırılmış Görüntü
plt.subplot(152)
plt.imshow(shifted_face2, cmap=plt.cm.gray)
plt.axis('off')

# 3. Döndürülmüş Görüntü
plt.subplot(153)
plt.imshow(rotated_face, cmap=plt.cm.gray)
plt.axis('off')

# 4. Kırpılmış Görüntü
plt.subplot(154)
plt.imshow(cropped_face, cmap=plt.cm.gray)
plt.axis('off')

# 5. Büyütülmüş Görüntü
plt.subplot(155)
plt.imshow(zoomed_face, cmap=plt.cm.gray)
plt.axis('off')

plt.subplots_adjust(wspace=0.05, left=0.01, bottom=0.01, right=0.99, top=0.99)  # Alt grafikleri sıkıştır
plt.show()
```
