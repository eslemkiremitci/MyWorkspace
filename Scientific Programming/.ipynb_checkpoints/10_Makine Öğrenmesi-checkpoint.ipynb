### **Makine Öğrenmesi:**
Makine öğrenmesi, insanın öğrenme yapısından esinlenerek geliştirilmiş ve yapay zekânın bir alt dalıdır. **Matematiksel ve istatistiksel** yöntemler kullanarak veri üzerinde çıkarımlar yapar ve bu çıkarımlar aracılığıyla **tahminlerde** bulunur. Amacı, bilgisayar sistemlerinin belirli görevleri programlamaya gerek kalmadan öğrenmesi ve geliştirmesidir.

### **Makine Öğrenmesi Günlük Hayatımızda Nerelerde Kullanılır?**
- **Spam Filtreleme:** E-postaları spam (istenmeyen) ve normal olarak sınıflandırır.
- **Web Aramaları:** Google gibi arama motorları, daha iyi arama sonuçları sunmak için makine öğrenmesi kullanır.
- **Posta Yönlendirme:** Postaların doğru yerlere ulaşmasını sağlamak için.
- **Dolandırıcılık Tespiti:** Bankacılık ve finans sistemlerinde dolandırıcılık faaliyetlerini belirler.
- **Film Önerileri:** Netflix gibi platformlar, kullanıcıların ilgisini çekebilecek içerikleri önerir.
- **Araç Asistanları:** Otonom sürüş sistemleri, trafik durumu ve sürüş güvenliği için veri işler.
- **Web Reklamcılığı:** Kullanıcı davranışlarına dayalı reklamlar gösterir.
- **Sosyal Ağlar:** İçerik önerileri ve kullanıcı bağlantılarını iyileştirir.
- **Konuşma Tanıma:** Siri veya Google Asistan gibi sistemler konuşmayı metne dönüştürür.

### **Makine Öğrenmesi Türleri**
1. **Denetimli Öğrenme (Supervised Learning):**
   - Veri setindeki **etiketli** verilerle çalışır.
   - **Sınıflandırma:** Veriyi önceden belirlenen kategorilere ayırır.
   - **Regresyon:** Sürekli bir çıktı tahmin eder (örneğin ev fiyatları).

2. **Denetimsiz Öğrenme (Unsupervised Learning):**
   - **Etiketsiz** veriler üzerinde çalışır.
   - **Kümeleme (Clustering):** Veriyi benzerliklerine göre gruplar.
   - **Birliktelik (Association):** Veriler arasındaki ilişkileri bulur.

3. **Yarı Denetimli Öğrenme (Semi-Supervised Learning):**
   - Hem etiketli hem de etiketsiz verilerle çalışır.
   - Daha az etikete ihtiyaç duyar.

4. **Takviyeli Öğrenme (Reinforcement Learning):**
   - Ödül ve ceza mekanizması ile öğrenir.
   - Amaç, belirli bir görevi en iyi şekilde tamamlamaktır.

### **<span style="background-color: lightblue;">Makine Öğrenmesinin Ana Türleri</span>**

1. **Sınıflandırma (Classification):**
   - **Tanımı:** Sınıflandırma, verilerin belirli kategorilere ayrılmasını sağlar. Hedef değişken ayrık (discrete) bir değere sahiptir.
   - **Örnekler:**
     - E-postaların spam olup olmadığının belirlenmesi.
     - Bir hastanın kanser olup olmadığının tahmini.
     - Kullanıcının alışveriş yapma olasılığının tahmini.
   - **Algoritmalar:**
     - Lojistik Regresyon (Logistic Regression)
     - Karar Ağaçları (Decision Trees)
     - K-En Yakın Komşu (K-Nearest Neighbors)(KNN)

2. **Regresyon (Regression):**
   - **Tanımı:** Sürekli (continuous) bir hedef değişkenin değerini tahmin eder. Sınıflandırmadan farkı, **tahmin edilen değerlerin sürekli olmasıdır** (örneğin sıcaklık, fiyat, gelir).
   - **Örnekler:**
     - Ev fiyatlarının tahmini.
     - Hisse senedi fiyatlarının tahmini.
     - Hava durumunun tahmini.
   - **Algoritmalar:**
     - Lineer Regresyon (Linear Regression)
     - Ridge ve Lasso Regresyon
     - Gradient Boosting

3. **Kümeleme (Clustering):**
   - **Tanımı:** Etiketlenmemiş (etiketsiz) verilerin gruplara ayrılmasıdır. Amaç, veri içerisindeki **doğal grupları veya desenleri bulmaktır.**
   - **Örnekler:**
     - Müşteri segmentasyonu.
     - Görüntü segmentasyonu.
   - **Algoritmalar:**
     - K-Means
     - Hiyerarşik Kümeleme
     - DBSCAN

4. **Takviyeli Öğrenme (Reinforcement Learning):**
   - **Tanımı:** Ödül ve ceza mekanizmasıyla çalışan bir öğrenme yöntemidir. Amaç, bir ajan (agent) için en iyi karar dizisini bulmaktır.
   - **Örnekler:**
     - Oyun oynayan yapay zeka (örneğin, satranç botları).
     - Robotik kontrol.

### **<span style="background-color: lightblue;">Karar Ağacı (Decision Tree) Nedir?</span>**
Karar ağacı, bir sınıflandırma veya regresyon probleminde, verilen verilerden anlamlı sonuçlar çıkarmak için kullanılan bir makine öğrenimi algoritmasıdır. Bu algoritma, bir dizi "kuralları" takip ederek bir problemin çözümüne ulaşır. Ağaç yapısında çalışır ve genellikle aşağıdaki bileşenlerden oluşur:

1. **Kök Düğüm (Root Node):**
   - Karar ağacının başlangıç noktasıdır.
   - Veri setindeki en önemli özelliği (değişkeni) seçerek, bu değişkene göre veri setini bölmeye başlar.
   
2. **Karar Düğümleri (Decision Nodes):**
   - Kök düğümden sonra gelen, veriyi belirli kurallara göre bölen düğümlerdir.
   - Her düğüm bir özelliği temsil eder ve bu özelliğe göre karar verir.

3. **Dallar (Branches):**
   - Her bir kararın sonucu olarak oluşan yolları temsil eder.
   - Örneğin, bir düğüm "Sıcaklık > 30°C" kuralını kontrol ediyorsa, "Evet" ve "Hayır" şeklinde iki dal oluşabilir.

4. **Yaprak Düğümleri (Leaf Nodes):**
   - Nihai karar noktalarıdır.
   - Sınıflandırma problemlerinde, her bir yaprak düğüm bir sınıf etiketi ile sonuçlanır.
   - Regresyon problemlerinde ise sayısal bir tahmin ile sonuçlanır.
     
### **<span style="background-color: lightblue;">Karar Ağacı Nasıl Çalışır?</span>**
1. **Veri Setini Bölme (Splitting):**
   - Karar ağacı, veri setini özelliklerine göre alt gruplara ayırır. Bu işlem, genellikle **bilgi kazancı (information gain)** veya **Gini indeksi** gibi ölçütlerle yapılır.

2. **Özellik Seçimi:**
   - Her düğümde en iyi özellik seçilir. Bu özellik, veri setini mümkün olduğunca iyi şekilde bölmelidir.
   - **Örneğin:** Bir hastalık teşhis problemi düşünelim. "Yaş", "Cinsiyet" ve "Ateş Durumu" gibi özellikler olabilir. İlk düğüm için en iyi ayırıcı özelliği seçerek ağaca başlanır.

3. **Karar Verme:**
   - Her dal bir "Evet/Hayır" veya "Doğru/Yanlış" kararını temsil eder. Ağaç, veri setini bu dallar boyunca takip eder ve nihai yaprak düğüme ulaşır.

4. **Sonuç:**
   - Yaprak düğüm, hangi sınıfa veya değere ait olduğunu belirtir.

### **Avantajları**
- **Kolay Anlaşılır ve Yorumlanabilir:**
  - Karar ağacındaki adımlar doğal bir şekilde insan mantığıyla uyumludur.
- **Hızlı Hesaplama:**
  - Basit algoritma yapısı sayesinde hızlı çalışır.
- **Kategorik ve Sürekli Verilerle Çalışabilir:**
  - Hem kategorik hem de sayısal veri ile kullanılabilir.

### **Dezavantajları**
- **Aşırı Uyum (Overfitting):**
  - Karar ağacı tüm veriyi çok iyi öğrenirse, yeni verilerde başarısız olabilir. Bu sorunu azaltmak için **budama (pruning)** işlemi uygulanabilir.
- **Dengesiz Veri Seti:**
  - Eğer veri setinde bazı sınıflar baskınsa, karar ağacı bu sınıfları daha fazla tercih edebilir.
- **Hassasiyet:**
  - Küçük veri değişiklikleri bile ağacın yapısını önemli ölçüde değiştirebilir.

### **<span style="background-color: lightblue;">Rastgele Orman (Random Forest)</span>**
Rastgele Orman, **birden fazla karar ağacını bir arada kullanan** ve oylama yaparak sonuç üreten bir makine öğrenimi algoritmasıdır. Karar ağacındaki **aşırı uyum (Overfitting)** problemini azaltır ve daha kararlı sonuçlar verir.

#### **Yapısı:**
- **Birden Fazla Karar Ağacı:**
  - Rastgele Orman, bir veri setini alır ve bu veri setinden rastgele örnekler seçerek birçok karar ağacı oluşturur.
- **Oylama (Voting):**
  - Her bir karar ağacı bağımsız olarak bir tahmin yapar ve bu tahminlerin **çoğunluk oyu** ile nihai sonuç belirlenir (sınıflandırma için).
  - **Regresyonda** ise ağaçların tahminlerinin **ortalaması** alınır.

#### **Avantajlar:**
- Aşırı uyum (overfitting) yapma ihtimali düşüktür.
- Daha kararlı ve doğru sonuçlar üretir.
- Büyük veri setleriyle iyi çalışır.

#### **Dezavantajlar:**
- Yorumlaması daha zordur (karar ağaçları kadar görsel değil).
- Hesaplama maliyeti yüksektir çünkü birden fazla ağaç oluşturulur.

## **<span style="background-color: lightblue;">Performans Ölçüm Metrikleri</span>**

###  **Karmaşıklık Matrisi (Confusion Matrix)**

Sınıflandırma problemlerinde modelin başarısını ölçmek için gerçek değerler ve tahmin edilen değerlere ait sayısal bilgileri ifade eden Karmaşıklık Matrisi (Confusion Matrix) adı verilen bir
matris oluşturulur.

Geliştirilen modelin performansının değerlendirilmesinde kullanılan metriklerden, doğruluk (accuracy), kesinlik (precision), duyarlılık (recall) ve f1-skor (f-score) hesaplanırken karmaşıklık
matrisi kullanılır.

![matris](matris.png)

#### **Doğruluk (Accuracy)**

Bir modelin ne kadar doğru tahmin yaptığını ölçer. Formülü:

![dogruluk](dogruluk.png)

- Doğru tahmin edilen pozitif ve negatif örneklerin toplamının tüm tahminlere oranıdır.
- **Genel bir performans ölçütüdür.**

**Dikkat:** Doğruluk, dengesiz veri kümelerinde yanıltıcı olabilir. Örneğin, 100 örnekten 95'i negatifse ve model hepsini negatif tahmin ederse doğruluk %95 olur ama bu yanıltıcıdır.

#### **Kesinlik (Precision)**

**Pozitif tahminlerin ne kadarının doğru olduğunu gösterir.** Formülü:

![kesinlik](kesinlik.png)

- Yanlış pozitiflerin (FP) etkisini azaltmayı amaçlar.
- Özellikle yanlış pozitif sonuçların maliyetinin yüksek olduğu durumlarda önemlidir.

**Örnek:** Spam filtrelerinde, kesinlik yüksekse, e-posta spam olarak işaretlenmişse gerçekten spam olma olasılığı yüksektir.

#### **Duyarlılık (Recall/Sensitivity)**

Gerçek pozitiflerin ne kadarını doğru tahmin ettiğimizi ölçer. Formülü:

![duyarlılık](duyarlılık.png)

- Yanlış negatiflerin (FN) etkisini azaltmayı amaçlar.
- Hasta tespiti gibi yanlış negatiflerin maliyetinin yüksek olduğu durumlarda önemlidir.


#### 5. **F1-Skor**

**Kesinlik ve duyarlılığın harmonik ortalamasıdır.** Dengeli bir değerlendirme yapar. Formülü:

![f1](f1.png)

- Kesinlik ve duyarlılık arasında bir denge kurar.
- Özellikle dengesiz veri setlerinde faydalıdır.




### **Regresyon Hataları**

Regresyon modellerinde kullanılan bazı hata metrikleri:
1. **Ortalama Mutlak Hata (MAE):**
   - Tahmin edilen ve gerçek değerler arasındaki farkların ortalaması.
2. **Ortalama Karesel Hata (MSE):**
   - Farkların karesinin ortalaması. Büyük hataları cezalandırır.
3. **Ortalama Karekök Hatası (RMSE):**
  - Tahmin edilen değerlerin gerçek değerlere uzaklığını ölçer.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier

data = pd.read_csv('heart_failure_clinical_records_dataset.csv')

data.head()
```

```python
data.isnull().sum() #Her sütünda kaç adet eksik veri var
```

```python
data.shape #verinin boyutlarını döndürür (satır, sütun)
```

```python
data.groupby('DEATH_EVENT').size() # sütununa göre gruplandırma yapar ve her grubun boyutunu (gruptaki eleman sayısını) döndürür.

```

```python
data.info() # DataFrame hakkında genel bilgi sağlar

```

```python
data.describe()

```

```python
y =data['DEATH_EVENT']
x=data.drop(['DEATH_EVENT'],axis=1) #axis 1 olunca sütun üzerinde işlem yapacam anlamına gelir. 0 olsa satır üzerinde işlem yapacam.

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=42)
```

```python


model = RandomForestClassifier(max_depth=3, random_state=42)

model.fit(x_train, y_train)
y_pred=model.predict(x_test)

print('Doğruluk: %.3f' % accuracy_score(y_test, y_pred))
print('Kesinlik: %.3f' % precision_score(y_test, y_pred))
print('Duyarlılık: %.3f' % recall_score(y_test, y_pred))
print('F1 Skor: %.3f' % f1_score(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

```python
model = DecisionTreeClassifier(max_depth=3, random_state=42)


model.fit(x_train, y_train)
y_pred=model.predict(x_test)

print('Doğruluk: %.3f' % accuracy_score(y_test, y_pred))
print('Kesinlik: %.3f' % precision_score(y_test, y_pred))
print('Duyarlılık: %.3f' % recall_score(y_test, y_pred))
print('F1 Skor: %.3f' % f1_score(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

```python
model=AdaBoostClassifier(n_estimators=100, algorithm="SAMME",)

model.fit(x_train, y_train)
y_pred=model.predict(x_test)

print('Doğruluk: %.3f' % accuracy_score(y_test, y_pred))
print('Kesinlik: %.3f' % precision_score(y_test, y_pred))
print('Duyarlılık: %.3f' % recall_score(y_test, y_pred))
print('F1 Skor: %.3f' % f1_score(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

```python

model = svm.SVC(kernel='linear')

model.fit(x_train, y_train)
y_pred=model.predict(x_test)

print('Doğruluk: %.3f' % accuracy_score(y_test, y_pred))
print('Kesinlik: %.3f' % precision_score(y_test, y_pred))
print('Duyarlılık: %.3f' % recall_score(y_test, y_pred))
print('F1 Skor: %.3f' % f1_score(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```

```python
model = KNeighborsClassifier(n_neighbors=5, weights= 'uniform', p=1)

model.fit(x_train, y_train)
y_pred=model.predict(x_test)

print('Doğruluk: %.3f' % accuracy_score(y_test, y_pred))
print('Kesinlik: %.3f' % precision_score(y_test, y_pred))
print('Duyarlılık: %.3f' % recall_score(y_test, y_pred))
print('F1 Skor: %.3f' % f1_score(y_test, y_pred))

confusion_matrix(y_test, y_pred)
```
